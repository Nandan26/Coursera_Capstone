# Web Scrapping using Beautiful Soup

```python
from bs4 import BeautifulSoup
import requests
```

```python
web = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(web,'lxml')
```

```python
print(soup.title) #title of webpage
```

#### display content of web page

```python
from IPython.display import display_html
table = str(soup.table) #get the table
display_html(table,raw=True) #display the table 
```

```python
import pandas as pd
```

```python
table_contents=[]
table=soup.find('table')
for row in table.findAll('td'):
    data = {}
    if row.span.text=='Not assigned':
        pass
    else:
        data['PostalCode'] = row.p.text[:3]
        data['Borough'] = (row.span.text).split('(')[0]
        data['Neighborhood'] = (((((row.span.text).split('(')[1]).strip(')')).replace(' /',',')).replace(')',' ')).strip(' ')
        table_contents.append(data)
```

```python
print(table_contents)
```

### Convert it into dataframe using pandas

```python
df=pd.DataFrame(table_contents)
df['Borough']=df['Borough'].replace({'Downtown TorontoStn A PO Boxes25 The Esplanade':'Downtown Toronto Stn A',
                                             'East TorontoBusiness reply mail Processing Centre969 Eastern':'East Toronto Business',
                                             'EtobicokeNorthwest':'Etobicoke Northwest','East YorkEast Toronto':'East York/East Toronto',
                                             'MississaugaCanada Post Gateway Processing Centre':'Mississauga'})
```

# For the Foursquare location, merging both the files

##### Reading the csv file conatining the latitudes and longitudes for various neighbourhoods in Canada and creating dataframe from it

```python
latti_longi = pd.read_csv('https://cocl.us/Geospatial_data')
latti_longi.head()
```

###### merging both the csv files

```python
latti_longi.rename(columns={'Postal Code':'PostalCode'},inplace=True)
df2 = pd.merge(df,latti_longi,on='PostalCode')
df2.head()
```

###### Unique values present in Borough

```python
df2['Borough'].unique()
```

###### Selecting all the borough which contains Toronto

```python
toronto = df2[df2['Borough'].str.contains('Toronto',regex=False)]
toronto
```

```python
import folium
```

###### Plotting all the points using folium

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(toronto['Latitude'], toronto['Longitude'], toronto['Borough'], toronto['Neighborhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```

# Explore and cluster the neighborhoods in Toronto

##### Kmeans Clustering for creating clusters

```python
from sklearn.cluster import KMeans
```

```python
k=5
toronto_clustering = toronto.drop(['PostalCode','Borough','Neighborhood'],1)
```

```python
kmeans = KMeans(n_clusters = k,random_state=0).fit(toronto_clustering)
kmeans.labels_
toronto.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
import numpy as np
```

```python
import matplotlib.cm as cm
import matplotlib.colors as colors
```

```python
# create map
map_clusters = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

# set color scheme for the clusters
x = np.arange(k)
ys = [i + x + (i*x)**2 for i in range(k)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, neighbourhood, cluster in zip(toronto['Latitude'], toronto['Longitude'], toronto['Neighborhood'], toronto['Cluster Labels']):
    label = folium.Popup(' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
